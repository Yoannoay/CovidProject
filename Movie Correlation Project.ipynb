```python
#Importing libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusting config of plots
```

```python
#Read in data

df = pd.read_csv(r'C:\Users\yoanc\OneDrive\Documents\Movies project\movies.csv')
df = df.dropna()
```

```python
#Looking at data
df.head()
```

```python
#checking for missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,percent_missing))

```

```python
# Data types for columns

df.dtypes
```

```python
#changing data type to clean data a lil

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')

df.head()
```

```python
#Correcting release year, some discrepancies exist

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
#sorting by highest grossing

df = df.sort_values(by = ['gross'], inplace=False, ascending=False)
```

```python
#setting it to default to showing all rows
#pd.set_option('display.max_rows', None)

pd.set_option('display.min_rows', None)
```

```python
df.head()
```

```python
# Scatter plot investigating the relationship between the budget of a film and it's gross revenue

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Film Budget')

plt.ylabel('Gross Earnings')
plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "red"}, line_kws = {"color":"blue"})
```

```python
# Further investigation of correlation

df.corr() #defaults to a pearson measure of correlation
```

```python
# Visualising correlation matrix

corr_matrix = df.corr(method = 'pearson')

sns.heatmap(corr_matrix, annot = True)

plt.title(' Correlation Matrix (Only numeric features)')

plt.xlabel('Variables')

plt.ylabel('Variables')

plt.show()
```

```python
# Assigning numeric values to all objects to be able to run further operations

df_numerised = df

for col_name in df_numerised.columns:
    if(df_numerised[col_name].dtype == 'object'):
        df_numerised[col_name] = df_numerised[col_name].astype('category')
        df_numerised[col_name] = df_numerised[col_name].cat.codes
        
df_numerised.head()
```

```python
df.head()
```

```python
df.head()
```

```python
# Correlation matrix for every field 

corr_matrix = df_numerised.corr(method = 'pearson')

sns.heatmap(corr_matrix, annot = True)

plt.title(' Correlation Matrix (Only numeric features)')

plt.xlabel('Variables')

plt.ylabel('Variables')

plt.show()
```

```python
#Making matrix easier to read using unstacking

correlation_mat = df_numerised.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
#Making matrix easier to read using pairs

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#Filtering through to only display those with a correlation higher than 0.5
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
